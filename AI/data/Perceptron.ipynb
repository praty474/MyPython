```python
import numpy as np
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # Inputs
y = np.array([0, 0, 0, 1])  # Outputs (labels)
```

```python
# Initialize weights, bias, learning rate, and epochs
weights = np.zeros(X.shape[1])  # Initialize weights as [0, 0]
bias = 0                       # Initialize bias as 0
learning_rate = 0.1            # Small value to control weight updates
epochs = 10  # Number of times to pass through the dataset
```

```python
# Activation function (Step function)
def step_function(z):
    return 1 if z >= 0 else 0
```

```python
print("Training the perceptron...\n")
for epoch in range(epochs):
    print(f"Epoch {epoch + 1}:")
    for i in range(len(X)):
        # Calculate the perceptron's output
        linear_output = np.dot(X[i], weights) + bias
        y_pred = step_function(linear_output)
        
        # Calculate the error
        error = y[i] - y_pred
        
        # Update weights and bias
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

        # Print updated weights and bias for understanding
        print(f"Input: {X[i]}, Target: {y[i]}, Predicted: {y_pred}")
        print(f"Updated Weights: {weights}, Updated Bias: {bias}\n")
```

```python
# Testing the perceptron
print("Testing the perceptron...\n")
for i in range(len(X)):
    linear_output = np.dot(X[i], weights) + bias
    y_pred = step_function(linear_output)
    print(f"Input: {X[i]}, Predicted Output: {y_pred}, Actual Output: {y[i]}")
```
